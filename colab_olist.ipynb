# Get the total delivery days

```python
#Average days and time for delivery
import pandas as pd
from datetime import datetime

order_dataset2['order_purchase_timestamp'] = pd.to_datetime(order_dataset2['order_purchase_timestamp'])
order_dataset2['order_delivered_customer_date'] = pd.to_datetime(order_dataset2['order_delivered_customer_date'])

order_dataset2["order_purchase_date"] = pd.to_datetime(order_dataset2['order_purchase_timestamp'].dt.strftime('%Y-%m-%d'))
order_dataset2['order_delivered_date'] = pd.to_datetime(order_dataset2["order_delivered_customer_date"].dt.strftime('%Y-%m-%d'))
order_dataset2["order_purchase_time"] = pd.to_datetime(order_dataset2['order_purchase_timestamp'].dt.strftime('%Y-%m-%d %H:%M:%S'))
order_dataset2['order_delivered_time'] = pd.to_datetime(order_dataset2["order_delivered_customer_date"].dt.strftime('%Y-%m-%d %H:%M:%S'))

order_dataset2['total_date_of_delivery'] = order_dataset2['order_delivered_date'].dt.date - order_dataset2["order_purchase_date"].dt.date #count days

#Filter the value to show sentence
mean_days = str(order_dataset2['total_date_of_delivery'].mean()).split('days')[0]
get_plus_time = str(order_dataset2['total_date_of_delivery'].mean()).split('days')[1]
adjust_time = get_plus_time.split('.')[0]
adjust_time
print(f'The average days that complete the order: {mean_days} days + time: {adjust_time}')
order_dataset2
```

# Accurate calculation for timelines (drop_duplicates)

```python
master_df_new = master_df.drop_duplicates(subset="order_id")
master_df_new
```

```python
# purchase to approve days
master_df_new['pur_appro_days'] = master_df_new['order_approved_at_date'] - master_df_new['order_purchase_date']

mapping_pur_appro_days = master_df_new['pur_appro_days'].astype(str).str.replace(' days','').to_dict()
master_df_new['pur_appro_days'] = master_df_new.index.map(mapping_pur_appro_days)
master_df_new['pur_appro_days'] = master_df_new['pur_appro_days'].astype(int)
master_df_new['pur_appro_days']
```

```python
# approve to carrier
master_df_new['appro_carrier_days'] = master_df_new['order_delivered_carrier_date_'] - master_df_new['order_approved_at_date']

mapping_appro_carrier_days = master_df_new['appro_carrier_days'].astype(str).str.replace(' days','').to_dict()
master_df_new['appro_carrier_days'] = master_df_new.index.map(mapping_appro_carrier_days)
master_df_new['appro_carrier_days'] = master_df_new['appro_carrier_days'].astype(int)
master_df_new['appro_carrier_days']
```

```python
# carrier to delivered
master_df_new['carrier_delivered_days'] = master_df_new['order_delivered_date'] - master_df_new['order_delivered_carrier_date_']

mapping_carrier_delivered_days = master_df_new['carrier_delivered_days'].astype(str).str.replace(' days','').to_dict()
master_df_new['carrier_delivered_days'] = master_df_new.index.map(mapping_carrier_delivered_days)
master_df_new['carrier_delivered_days'] = master_df_new['carrier_delivered_days'].astype(int)
master_df_new['carrier_delivered_days']
```

```python
#Check shipping limited time (seller pass order to carrier)
master_df_new['shipping_limit_date'] = pd.to_datetime(master_df_new['shipping_limit_date'])

master_df_new['seller_to_carrier_date'] = pd.to_datetime(master_df_new['shipping_limit_date'].dt.strftime('%Y-%m-%d'))
master_df_new['seller_to_carrier_date']
```

```python
master_df_new['seller_to_carrier_intervals'] = master_df_new['seller_to_carrier_date'] - master_df_new['order_delivered_carrier_date_']
mapping_seller_to_carrier = master_df_new['seller_to_carrier_intervals'].astype(str).str.replace(' days','').to_dict()
master_df_new['seller_to_carrier_intervals'] = master_df_new.index.map(mapping_seller_to_carrier)

#Change to int
master_df_new['seller_to_carrier_intervals'] = master_df_new['seller_to_carrier_intervals'].astype(int)
master_df_new['seller_to_carrier_intervals'].describe()
```
